```python
import pandas as pd
import yfinance as yf

# pick start and end dates
start = '2010-01-01'
end = '2025-12-31'

# pick stocks 
stock1_string = 'spy'
stock2_string = 'spxl'
stock3_string = 'qqq'
stock4_string = 'tqqq'

# Fetching data for stock1
stock1 = yf.Ticker(stock1_string.lower())
stock1_data = stock1.history(interval='1d',start=start, end=end)
stock1_data['Daily % Change'] = stock1_data['Close'].pct_change().fillna(0)

# Fetching data for stock2
stock2 = yf.Ticker(stock2_string.lower())
stock2_data = stock2.history(interval='1d',start=start, end=end)
stock2_data['Daily % Change'] = stock2_data['Close'].pct_change().fillna(0)

# Fetching data for stock1
stock3 = yf.Ticker(stock3_string.lower())
stock3_data = stock3.history(interval='1d',start=start, end=end)
stock3_data['Daily % Change'] = stock3_data['Close'].pct_change().fillna(0)

# Fetching data for stock2
stock4 = yf.Ticker(stock4_string.lower())
stock4_data = stock4.history(interval='1d',start=start, end=end)
stock4_data['Daily % Change'] = stock4_data['Close'].pct_change().fillna(0)

# Intraday range
stock1_data['Intraday range'] = stock1_data['High'] - stock1_data['Low']
stock2_data['Intraday range'] = stock2_data['High'] - stock2_data['Low']
stock3_data['Intraday range'] = stock3_data['High'] - stock3_data['Low']
stock4_data['Intraday range'] = stock4_data['High'] - stock4_data['Low']

# $ Volume
stock1_data['$ Volume'] = stock1_data['Close'] * stock1_data['Volume']
stock2_data['$ Volume'] = stock2_data['Close'] * stock2_data['Volume']
stock3_data['$ Volume'] = stock3_data['Close'] * stock3_data['Volume']
stock4_data['$ Volume'] = stock4_data['Close'] * stock4_data['Volume']

# Row Number
stock1_data['Row Number'] = [i + 1 for i in range(len(stock1_data))]
stock2_data['Row Number'] = [i + 1 for i in range(len(stock2_data))]
stock3_data['Row Number'] = [i + 1 for i in range(len(stock3_data))]
stock4_data['Row Number'] = [i + 1 for i in range(len(stock4_data))]

# Create Last Close column
stock1_data['Last Close'] = 'No'
stock2_data['Last Close'] = 'No'
stock3_data['Last Close'] = 'No'
stock4_data['Last Close'] = 'No'

stock1_data.loc[stock1_data.index[-1],'Last Close'] = 'Yes'
stock2_data.loc[stock2_data.index[-1],'Last Close'] = 'Yes'
stock3_data.loc[stock3_data.index[-1],'Last Close'] = 'Yes'
stock4_data.loc[stock4_data.index[-1],'Last Close'] = 'Yes'

# assign the first row to First for calculating period return in Domo
stock1_data.loc[stock1_data.index[0],'Last Close'] = 'First'
stock2_data.loc[stock2_data.index[0],'Last Close'] = 'First'
stock3_data.loc[stock3_data.index[0],'Last Close'] = 'First'
stock4_data.loc[stock4_data.index[0],'Last Close'] = 'First'

# Reformat both indexes
stock1_data.index = pd.to_datetime(stock1_data.index).strftime('%Y-%m-%d')
stock2_data.index = pd.to_datetime(stock2_data.index).strftime('%Y-%m-%d')
stock3_data.index = pd.to_datetime(stock3_data.index).strftime('%Y-%m-%d')
stock4_data.index = pd.to_datetime(stock4_data.index).strftime('%Y-%m-%d')

# Add ticker column
stock1_data['Ticker'] = stock1_string.upper()
stock2_data['Ticker'] = stock2_string.upper()
stock3_data['Ticker'] = stock3_string.upper()
stock4_data['Ticker'] = stock4_string.upper()

# concat to one df and save to csv for uploading to Domo
df = pd.concat([stock1_data,stock2_data,stock3_data,stock4_data])

df.to_csv('3XLeverage.csv')
df.tail()
```

```python
import pandas as pd

# Assuming stock1_data is your DataFrame with columns 'Date' and 'Close' and 'Date' is set as the index

# Convert index to datetime if it's not already
stock1_data.index = pd.to_datetime(stock1_data.index)
stock2_data.index = pd.to_datetime(stock2_data.index)
stock3_data.index = pd.to_datetime(stock3_data.index)
stock4_data.index = pd.to_datetime(stock4_data.index)

# Extract year from the index
stock1_data['Year'] = stock1_data.index.year
stock2_data['Year'] = stock2_data.index.year
stock3_data['Year'] = stock3_data.index.year
stock4_data['Year'] = stock4_data.index.year

# Group by 'Year' and get the first and last closing prices for each year
yearly_closes_1 = stock1_data.groupby('Year')['Close'].agg(['first', 'last'])
yearly_closes_2 = stock2_data.groupby('Year')['Close'].agg(['first', 'last'])
yearly_closes_3 = stock3_data.groupby('Year')['Close'].agg(['first', 'last'])
yearly_closes_4 = stock4_data.groupby('Year')['Close'].agg(['first', 'last'])

# Calculate the percent return for each year
yearly_returns_1 = (yearly_closes_1['last'] - yearly_closes_1['first']) / yearly_closes_1['first']
yearly_returns_2 = (yearly_closes_2['last'] - yearly_closes_2['first']) / yearly_closes_2['first']
yearly_returns_3 = (yearly_closes_3['last'] - yearly_closes_3['first']) / yearly_closes_3['first']
yearly_returns_4 = (yearly_closes_4['last'] - yearly_closes_4['first']) / yearly_closes_4['first']

# convert to pandas df and name Annual Return column
yearly_returns_1 = pd.DataFrame(yearly_returns_1, columns=['Annual Return'])
yearly_returns_2 = pd.DataFrame(yearly_returns_2, columns=['Annual Return'])
yearly_returns_3 = pd.DataFrame(yearly_returns_3, columns=['Annual Return'])
yearly_returns_4 = pd.DataFrame(yearly_returns_4, columns=['Annual Return'])

# Add ticker column
yearly_returns_1['Ticker'] = stock1_string.upper()
yearly_returns_2['Ticker'] = stock2_string.upper()
yearly_returns_3['Ticker'] = stock3_string.upper()
yearly_returns_4['Ticker'] = stock4_string.upper()

# concat into one df and save to csv for uploading to Domo
yearly_returns = pd.concat([yearly_returns_1, yearly_returns_2, yearly_returns_3, yearly_returns_4])
yearly_returns.to_csv('Yearly Returns.csv')

print(yearly_returns.head())
print(yearly_returns.tail())
```
